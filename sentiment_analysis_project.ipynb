```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('/content/3) Sentiment dataset.csv')

# Show first 5 rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Check sentiment categories
df['Sentiment'].value_counts()
```

```python
# Top 10 sentiment categories

top_sentiments = df['Sentiment'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    x=top_sentiments.index,
    y=top_sentiments.values
)

# Rotate labels
plt.xticks(rotation=45)

# Titles
plt.title('Top 10 Sentiment Categories')

plt.xlabel('Sentiment')
plt.ylabel('Count')

plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Combine all text

text = ' '.join(df['Text'].astype(str))

# Create word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(text)

# Plot word cloud
plt.figure(figsize=(12,6))

plt.imshow(wordcloud, interpolation='bilinear')

plt.axis('off')

plt.title('Word Cloud of Sentiment Dataset')

plt.show()
```

```python
# Sentiment percentages

sentiment_percent = df['Sentiment'].value_counts(normalize=True) * 100

print(sentiment_percent)
```
